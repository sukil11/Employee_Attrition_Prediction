```python
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
import pandas as pd
```

```python
df.info()
```

```python
df["Attrition"]=df["Attrition"].map({'No':0,'Yes':1})
df["Gender"]=df["Gender"].map({'Male':0,'Female':1})
df["OverTime"]=df["OverTime"].map({'No':0,'Yes':1})
```

```python
df = pd.get_dummies(df, columns=[ 'BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus'], drop_first=True)
```

```python
df.head()
```

```python
df.columns
```

```python
# Features and target variable
x = df[['Age', 'DailyRate', 'DistanceFromHome', 'Education', 
        'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 
        'JobLevel', 'JobSatisfaction', 'MonthlyIncome', 
        'NumCompaniesWorked', 'OverTime']]  # After one-hot encoding, include relevant columns
y = df["Attrition"]
```

```python
# Split the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Resample the data using SMOTE
smote = SMOTE(random_state=42)
x_res, y_res = smote.fit_resample(x_train, y_train)
```

```python
# Scale the features after splitting (to prevent data leakage)
scaler = StandardScaler()
x_res = scaler.fit_transform(x_res) 
x_test = scaler.transform(x_test)
```

```python
# Random Forest
rf = RandomForestClassifier(random_state=42, max_depth=15)
rf.fit(x_res, y_res)
y_pred_rf = rf.predict(x_test)

# Decision Tree
dt = DecisionTreeClassifier(max_depth=15)
dt.fit(x_res, y_res)
y_pred_dt = dt.predict(x_test)

# Evaluate models
def evaluate_model(y_test, y_pred, model_name):
    print(f"Evaluation for {model_name}:")
    print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
    print(f"Precision: {precision_score(y_test, y_pred)}")
    print(f"Recall: {recall_score(y_test, y_pred)}")
    print(f"F1-Score: {f1_score(y_test, y_pred)}")
    print(f"AUC-ROC: {roc_auc_score(y_test, y_pred)}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

# Evaluate Random Forest model
evaluate_model(y_test, y_pred_rf, "Random Forest")

# Evaluate Decision Tree model
evaluate_model(y_test, y_pred_dt, "Decision Tree")
```

```python
import joblib
joblib.dump(rf, 'random_forest_model.joblib')
joblib.dump(dt, 'decision_tree_model.joblib')
joblib.dump(scaler, 'scaler.joblib')
```
